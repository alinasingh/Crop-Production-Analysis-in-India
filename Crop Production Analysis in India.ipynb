# Visualization on the basis of Area Status

```python
df["Area"].nunique()
```

```python
df["Area"].max()
```

```python
df["Area"].value_counts().head()
```

```python
df["Area"].value_counts().tail()
```

```python
plt.figure(figsize=(5,4))
plt.boxplot(df.Area)
plt.show()
```

## Crop distribution according to Area status:

```python
crop = df.groupby('Crop')['Area'].sum().reset_index().sort_values('Area', ascending=False).head(15)
plt.figure(figsize=(5,4))
ax=sns.barplot(crop.Crop, crop.Area)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.yscale('log')
plt.title('Overall Crops distribution by Area')
plt.show()
```

## Season distribution according to Area status:

```python
season_a = df.groupby('Season')['Area'].sum().sort_values(ascending=False).head(10)
season_a
```

```python
season_a = df.groupby('Season')['Area'].sum().head(10)
plt.figure(figsize=(5,4))
ax=sns.barplot(x='Season', y='Area', data= df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.yscale('log')
plt.title('Seasonal_Crops according to Area distribution')
plt.show()


plt.figure(figsize=(6,4))
sns.boxplot(x="Season", y="Area" , data=df, width=0.66, saturation=0.65 )
plt.xticks(rotation=90, ha="right")
plt.yscale('log')
plt.show()
```

## Crop_year distribution according to Area status:

```python
plt.figure(figsize=(6,5))
ax=sns.barplot(x='Crop_Year', y='Area', data= df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.yscale('log')
plt.xlabel('Crop_year')
plt.title('Crop_Year according to Area distribution')
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.lineplot(x='Crop_Year', y='Area', data= df, marker="o")
plt.xticks(rotation=45, ha="right")
plt.xlabel('Crop_year')
plt.yscale('log')
plt.title('Crop_Year according to Area distribution')
plt.show()
```

## Season distribution according to Production Status:

## Crop distribution according to Production status:

```python
crop = df.groupby('Crop')['Production'].sum().reset_index().sort_values('Production', ascending=False).head(10)
plt.figure(figsize=(7,6))
ax=sns.barplot(crop.Crop, crop.Production)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.yscale('log')
plt.title('Overall Crops distribution by Production')
plt.show()
```

## Crop_Year according to Production distribution

```python
season = df.groupby('Crop_Year')['Production'].sum().head(10)
plt.figure(figsize=(8,5))
ax=sns.barplot(x='Crop_Year', y='Production', data= df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.yscale('log')
plt.title('Crop_Year according to Production distribution')
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.lineplot(x='Crop_Year', y='Production', data= df, marker="o", color="red")
plt.xticks(rotation=45, ha="right")
plt.xlabel('Crop_year')
plt.title('Crop_Year according to Production distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
pvt1=pd.pivot_table(data=df, index="State_Name", columns="Season",aggfunc={"Production":np.mean}, fill_value=0).head()
sns.heatmap(pvt1, annot=True, fmt="2f" )
plt.title("Heat map")
plt.show() 
```
